```python
# If the input are not linearly separable then we need to use polynomial Feature( not column just method)

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
n=pd.read_csv("pokemon.csv")
n.head(2)
```

```python
sns.scatterplot(x="Attack",y="Speed",hue="Speed",data=n)
plt.show()
```

```python
x=n[["Attack","Speed"]]
y=n["Legendary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

# Implementing Polynomial Features

```python
x=n[["Attack","Speed"]]
y=n["Legendary"]
```

```python
# Transforming the data 

from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=4)
pf.fit(x)
x = pd.DataFrame(pf.transform(x))
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
