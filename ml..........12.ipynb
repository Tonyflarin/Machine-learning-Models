```python
#polynomial regression (which refers to a CURVE)

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
n=pd.read_csv("imdb.csv")
n.head(2)
```

```python
#to check the co-relation

n[["Metascore","IMDB Rating"]].corr()
```

```python
for i in n.columns:
    n[i].fillna(n[i].mode()[0],inplace=True)
```

```python
plt.scatter(n["Metascore"],n["IMDB Rating"])
plt.xlabel("Metascore")
plt.ylabel("IMDB Rating")
plt.show()
```

```python
x=n[["Metascore"]]
y=n[["IMDB Rating"]]
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=4)
x=pf.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#To check its accuracy

lr.score(x_test,y_test)*100
```

```python
y_pred=lr.predict(x)
```

```python
plt.scatter(n["Metascore"],n["IMDB Rating"])
plt.plot(n["Metascore"],y_pred,c="r")
plt.show()
```
