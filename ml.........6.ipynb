```python
#Scaling refers to reduce the numbers of larger value to smaller value but which does not change the nature 
#(both the distplot will be same)

#it involves two methods:
# 1.standardlization scaling
# 2.normalization (minmaxscaling)
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
n=pd.read_csv("asd.csv")
n.head(3)
```

```python
n.describe()
```

```python
n.isnull().sum()
```

```python
sns.distplot(n["Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)"])
plt.show()
```

```python
n["Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)"].fillna(n["Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)"].mode()[0],inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
n["ss_1year GPA"]=ss.fit_transform(n[["Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)"]])
```

```python
n.head(5)
```

```python
n.describe()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(n["ss_1year GPA"])
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(n["Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)"])


plt.subplot(1,2,2)
plt.title("after")
sns.distplot(n["ss_1year GPA"])
plt.show()
```


# normalization(minmaxscaling)




```python
m=pd.read_csv("asd.csv")
m.head(3)
```

```python
m.describe()
```

```python
sns.distplot(m["Your Matric (grade 12) Average/ GPA (in %)"])
plt.show()
```

```python
m.isnull().sum()
```

```python
m["Your Matric (grade 12) Average/ GPA (in %)"].fillna(m["Your Matric (grade 12) Average/ GPA (in %)"].mean(),inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
m["grade_12_minmax"]=ms.fit_transform(m[["Your Matric (grade 12) Average/ GPA (in %)"]])
```

```python
m.head(3)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(m["Your Matric (grade 12) Average/ GPA (in %)"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(m["grade_12_minmax"])
plt.show()
```
