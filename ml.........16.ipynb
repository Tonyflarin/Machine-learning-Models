# Logistic Regression

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
n=pd.read_csv("pokemon.csv")
n.head(2)
```

```python
sns.scatterplot(x="Speed",y="Generation",data=n)
plt.show()
```

```python
x=n[["Speed"]]
y=n[["Generation"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#Poor score so the prediction score will be also poor (when the score is very poor then the prediction score will be also poor)

lr.score(x_test,y_test)*100
```

```python

lr.predict([[45]])
```

```python
#If we see the scatterplot (75-175) the generation is 1 , (50-60) the generation is 5 , (25-50) the generation is 3 , (1-10) its 2 

sns.scatterplot(x="Speed",y="Generation",data=n)
sns.lineplot(x="Speed",y=lr.predict(x),data=n,color="red")
plt.show()
```

# Logistic Regression - Multiple input


```python
m=pd.read_csv("pokemon.csv")
m.head(2)
```

```python
sns.scatterplot(x="Speed",y="Generation",data=m,hue="Legendary")
plt.show()
```

```python
x=m.iloc[:,4:11]
y=m["Generation"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[45,45,67,89,90,87,56]])
```
