# Working with Imbalancing data's

```python
# Equalizing all the Imbalancing datas in a column and Balancing the data's

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
n=pd.read_csv("pokemon.csv")
n.tail(30)
```

```python
n["Type 2"].fillna(n["Type 2"].mode()[0],inplace=True)
```

```python
n.isnull().sum()
```

```python
n["Legendary"].value_counts()
```

```python
x=n.iloc[:,4:12]
y=n["Legendary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#It majority of the cases it won't show the accurate value in such cases 

lr.predict([[600,80,110,120,130,90,70,6,]])
```

# Implementing imblearn - Under Sampling

```python
#Random values of majority data in a column

from imblearn.under_sampling import RandomUnderSampler 
```

```python
ru =RandomUnderSampler()
ru_x ,ru_y= ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ru_x,ru_y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#Now the prediction value will more accurate

lr.predict([[600,80,110,120,130,90,70,6,]])
```

# Implementing imblearn - Over Sampling

```python
# copies of minority data in a column to equalize 

from imblearn.over_sampling import RandomOverSampler
ro=RandomOverSampler
ro_x ,ro_y= ru.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ro_x,ro_y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#Now the prediction value will more accurate

lr.predict([[600,80,110,120,130,90,70,6,]])
```
