```python
#multiple linear regression 
#two or more input
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
n=pd.read_csv("imdb.csv")
n.head(3)
```

```python
n.describe()
```

```python
n["Metascore"].fillna(n["Metascore"].mean(),inplace=True)
```

```python
#To check the Reletion 
sns.pairplot(data=n)
plt.show()
```

```python
#corr() is used to find the co-releation between the two features

sns.heatmap(n[["Duration","IMDB Rating","Metascore"]].corr(),annot=True)
plt.show()
```

```python
x=n[["Duration","Metascore"]]
y=n[["IMDB Rating"]]
```

```python
#to check the deminision
x.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#To check the score wether it is good or bad (still 48.416 is not a good score)

lr.score(x_test,y_test)*100
```

```python
#For multiple linear regression formula
#y= m1x1 + m2x2 + m3x3 +c

#m
lr.coef_
```

```python
#c
lr.intercept_
```
